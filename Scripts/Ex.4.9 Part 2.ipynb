# 3. You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

```python
df_final_dataframe['order_hour_of_day'].plot.hist(bins = 25)
```

# describe what the histogram demonstrates
The histogram shows that most orders are placed during late morning to early evening, with a clear peak between 10 AM and 4 PM. Very few orders occur during the early morning hours (midnight–6 AM). This suggests that customer activity is strongly concentrated during typical daytime hours, likely aligned with daily routines such as work breaks and afternoon shopping.

# 4. The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
df_final_dataframe['loyalty_flag'].value_counts().plot.bar()
```

# Description of Bar chart
The bar chart indicates that regular customers form the largest group, followed by loyal customers, while new customers represent the smallest segment. This suggests that Instacart retains a substantial base of repeat users, but there is still potential to convert regular customers into loyal ones through targeted marketing strategies.

# 6. Now that you have information about customers, you need to conduct some exploratory analysis of customer demographics to inform the targeted marketing campaigns. First, determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents

```python
# sampling taking 100k random rows from dataframe to avoid memory error
df_2 = df_final_dataframe.sample(n = 50000, random_state = 4)
```

```python
line_2 = sns.lineplot(data = df_2, x = 'age', y = 'n_dependents')
```

# Description of line chart
The line chart shows a weak and relatively stable relationship between age and number of dependents. Across different age groups, the average number of dependents remains fairly consistent, with only minor fluctuations. This suggests that age alone is not a strong predictor of family size within this customer base.

# 7. You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.

```python
sns.scatterplot(x = 'age', y = 'income',data = df_final_dataframe)
```

# Description of scatter plot
The scatter plot indicates that customers aged 20 to around 40 generally show incomes clustered at lower to mid levels, while customers aged 45 to 80 tend to display higher income values, with incomes extending up to 600,000. This suggests that income appears to increase with age, likely reflecting career progression and accumulated earnings over time. However, there is still considerable variation within each age group, indicating that age is not the sole factor influencing income.
