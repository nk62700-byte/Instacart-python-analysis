# Confirming the results of the merge using the merge flag.

```python
#I merged the orders dataset with the products dataset using product_id to add product details to each order.
merged_df = pd.merge(orders_products_combined, df_prods, on = 'product_id', how = 'left', indicator='Match')
```

```python
# dropping merge column in order_products_combined 
orders_products_combined = orders_products_combined.drop(['_merge'], axis=1)
```
