# To avoid memory issues, let’s work with a subset of the dataframe for the first one million rows.

```python
# create a subset
df = df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
# creating the “price_label” and “busiest_day” columns
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# I created a new list that labels each order’s day of the week as “Busiest days”, “Slowest days”, or “Regular days” based on the value in orders_day_of_week.
results = []                              
for value in df_ords_prods_merge["orders_day_of_week"]:              
  if value in (0,1):
    results.append("Busiest days")
  elif value in (3,4):
    results.append("Slowest days")
  else:
    results.append("Regular days")
```

```python
results
```

```python
df_ords_prods_merge['busiest_days'] = results
```

# 4. The senior technical officer at Instacart wants you to identify the busiest hours of the day.

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

# The busiest shopping hours are between 10:00–16:00, with peak orders around 10–15. Orders are lowest late night and early morning (0:00–6:00), while the rest of the day shows average ordering activity.

```python
results_hours = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in (9, 10, 11, 12, 13, 14, 15, 16):
        results_hours.append("Most orders")
    elif value in (23,6, 0, 1, 5, 2, 4, 3):
        results_hours.append("Fewest orders")
    else:
        results_hours.append("Average orders")
```

```python
results_hours
```

```python
df_ords_prods_merge['busiest_period_of_day'] = results_hours
```
